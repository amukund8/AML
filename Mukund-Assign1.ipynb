1: Define each of the following machine learning terms in YOUR OWN WORDS:

	a. Training Dataset: The input/data with features that the machine learning model uses to learn/train
    Testing dataset: The data/features that the machine learning model uses to evaluate its performance
    Validation Dataset: THe data/features that the machine learning model uses to calculate its optimal performance based on selecting optimal hyperparameters

	b. Ground truth: A feature of the data that is the absolute truth that is used to compare against the predicted value that is calculated by a ML model on test data
    Label: The target in a machine learning model that the model calculates loss from.

    c. Pre-processing: Fixing the data to remove inaccuracies, incorrect information, converting into correct data types, and removing unwanted features so that the processed data can be correctly used as input to the ML model
    feature: If the data is tabular, where the rows are the sample numbers, then the columns are the data's features
    numerical: Variables that have a number value that falls within a certain range
    Nominal: Variables that have a category that cannot perform mathematical operations on them, such as a gender or type of cancer

    d. Decision Surface: When doing a classification problem, there is a hyperplane that exists in some space that differentiates data points from one class to another. ANN's solve for them through stochastic gradient descent as a solution to the optimal solution problem that exists in classification.  
    Model Validation: The process of evaluating the model's hyperparameters through examination of its performance. This can be done in different ways
    Accuracy: A metric that is used in model evaluation that measures how many correctly predicted samples over the total number of samples
    Cross-validation: If validation is measuring the model's performance based on the testing data, cross-validation is using variations of the testing data to ensure and cross check the moel's performance
    Parameters: In the case of a neural network they are the weights that are updated each time a model backpropogates. In more general terms, they are the variables that are optimized for the testing phase
    Hyperparameters: The design choices that you make in a machine learning model that do not change during training, but play a significant role in the model's perforamance
	Overfit: When the model displays a low loss/converges very accurately on the training data, but has very poor performance on the testing data. This occurs when the data fails to capture the underlying patterns of the training data. 
	e. Reclassificaiton: When you use the training dataset for testing, and can be used to indicate problems with the code or see how the learning is working for the model

Question 6: Classify iris dataset from question 4 using Naive Bayes and decisoin trees. Divide dataset into two portiosn, one for training and the other for tesitng using test_train_split (and make sure shuffle is on)

```python

from sklearn.metrics import accuracy_score
from sklearn.naive_bayes import GaussianNB
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split

X, y = load_iris(return_X_y=True)
# 50% training set, 50% testing set 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=0)

# Naive Bayes Classifier
gnb=GaussianNB()
y_pred = gnb.fit(X_train, y_train).predict(X_test)
accuracy_nb=accuracy_score(y_test,y_pred)


# Decision Tree Classifier
dt=DecisionTreeClassifier()
dt.fit(X_train,y_train)
y_pred_dt=dt.predict(X_test)
accuracy_dt=accuracy_score(y_test,y_pred_dt)
```

```python
print(f'Naive Bayes Classifier Accuracy w/ 50% training set: {accuracy_nb}')
print(f'Decision Tree Classifier Accuracy w/ 50% training set: {accuracy_dt}')
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)

# Naive Bayes Classifier
gnb=GaussianNB()
y_pred = gnb.fit(X_train, y_train).predict(X_test)
accuracy_nb=accuracy_score(y_test,y_pred)


# Decision Tree Classifier
dt=DecisionTreeClassifier()
dt.fit(X_train,y_train)
y_pred_dt=dt.predict(X_test)
accuracy_dt=accuracy_score(y_test,y_pred_dt)
```

```python
print(f'Naive Bayes Classifier Accuracy w/ 75% training set: {accuracy_nb}')
print(f'Decision Tree Classifier Accuracy w/ 75% training set: {accuracy_dt}')
```

```python
test_list=[.01,0.05,0.10,0.25,0.50,0.75,0.90,0.95,0.99]
accuracies=[]
for i in test_list:
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=i, random_state=0)

    # Naive Bayes
    gnb=GaussianNB()
    y_pred = gnb.fit(X_train, y_train).predict(X_test)
    accuracy_nb=accuracy_score(y_test,y_pred)


    # Decision Tree Classifier
    dt=DecisionTreeClassifier()
    dt.fit(X_train,y_train)
    y_pred_dt=dt.predict(X_test)
    accuracy_dt=accuracy_score(y_test,y_pred_dt)

    accuracies.append([accuracy_nb,accuracy_dt])
    print(f'Naive Bayes Classifier Accuracy w/ {1-i} training set: {accuracy_nb}')
    print(f'Decision Tree Classifier Accuracy w/ {1-i} training set: {accuracy_dt}')
```

```python
# Create some sample data
x = test_list
y1 = [i[0] for i in accuracies]
y2 = [i[1] for i in accuracies]

# Plot the first line
plt.plot(x, y1, label='Accuracy of Naive Bayes Classifier for different testing sizes')

# Plot the second line on the same graph
plt.plot(x, y2, label='Accuracy of Decision Tree Classifier for different testing sizes')

# Add labels and a legend
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Accuracies across different testing sizes ')
plt.legend()

# Show the plot
plt.show()
```

Question 6 Answers:

a. Shuffle is extremely important for this dataset, since the data is originally grouped by class as shown above. It is important to shuffle the dataset with the train_test_split function (whose default is shulffle=True) in order to mix the data so that it trains properly. Otherwise the split will occur only on data from a singular or binary class rather than having a mix of all three classes. The data will become overfit if it is not shuffle properly. 

b. Based on the statistics of accuracies based on varying testing sizes, both models have similar performance, but on average the decision tree classifier has a higher mean, which can also be seen in the graph above. The standard deviation is slightly more with the Naive Bayes Classifier, which means that the decision tree classifier is a bit more stable as well. For these two reasons, we can conclude that the decision tree classifier has a higher performance.

c. Yes, indeed, more training does help the classifier, as is seen from the graph above. As the testing size increases (less training), the accuracy has a large drop off, and can even dip to 38% roughly for both classifiers. For this reason, it is extremely important to divide your training and testing splits to be majority testing, minority training, and the cutoff for this point can be seen at 90% training, 10% testing.

d. The performance does indeed plateau, as there is a point at 95% training, 5% testing that the performance results in 100% accuracy for both classifiers. At 100% accuracy, there is no room for improvement, which signifiies the ideal training and testing split for the iris dataset.
