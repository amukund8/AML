Percpetron: The textbook defines a perceptron as a set of artificial neurons that use a randomized set of weights and a linear basis function calculate a value that goes through a non linearity. The randomized set of weights are updated using gradient descent across a loss funciton that measures the difference between the true label and linear basis value that goes through the non linearity. The rate at which the weights are updated are based on the learnign rate, eta. The purpose of gradient descent is to serve as a means to solve the optimization problem. THe loss or cost function, in the textbook's case is the sum of square errors between the predicted and true label. The sum of square errors is exactly what it soudns like, which is the sum of true-predicted. Some disadvantages to the perceptron algorithm are that a learning rate that is too high may overshoot a global minimum, while a learning rate too small may lead to a solution that is stuck in a local minimum, requiring the need for more epochs. Some advantages of the perceptron allow for the plotting of a decision regions of the perceptron and view the seperation. However, perceptrons do not work on data that are not easily linearly seperable, but there do exist other classifiers such as SVM that do not require teh data to be completely linearly seperable. Perceptron are the basis of multi layer percpetrons, and the logic of optimization via gradient descent is so strong that it is used in deep neural networks such as CNNs, transformers, and has functionality in unsupervised learning as well. This also makes them quite robust. Pruning may need to be done on the nodes with low weight.

SVM: According to the textbook, while perceptrons minimize misclassification errors, the purpose of an SVM is to maximize the distance boundaries between different classes. Support vectors are the points that are closest to these decision boundaries. The objective funciton of an SVM is the is the maximization of the decision boundaries, represented by two equations which are a negative and positive hyperplane, where all negative classes fall on one side of the negative hyperplane and vise versa for the other one. The slack variable is what allows for variables that requiring complete linear seperability so that optimization can be achieved. Kernalizing an SVM can help achieve the ssame result. SVMs are slightly more complicated to other algorithms such as logistic regression, which make them a little more complex to implement. Interestingly, a simple perceptron cannot solve a XOR problem, since linear seperability does not exist, but a kernalized SVM can. The model can be considered robust, due to capability, while yielding strong accuaracy.

Decision Trees: The main advantage of decision trees that we can uses the concept of information gain to make a decision, which aids interpretability. Information gain is done through an objection function that splits a node based on an impurity score of the parent and the jth child node. The impurity score is calculated by entropy, which measures the ratio of examples that belong to a certain class for the current leaf node. The maximized value is when everything belongs to the same class. By maximizing the gini impurity, the probability of misclassification is minimiized. This impurity score The process splits until all child nodes are the same class. Decision trees allow for the visualization of each split, and from a classification perspective, the amount of information gain can be traced down the tree. The more important classes can be traced from teh root, while the lesser important classes can be seen in the leaves. SOme cons of individual decision trees are that they suffer from high variance, which can be tackled through the use of a random forest model. An accurate decision tree may often need additional pruning, just as pruning may be required in perceptrons with nodes that contribute to low weight.

Random Forest: Random forest is an ensemble method, which combines the outputs of multiple models to come up with a prediction, which proves stronger than using an individual mdoel. In this case, it combines an ensemble of decision trees. After choosing a number of decision trees based on choosing n examples, we select a select number of random features and use the information gain to split. By looking at the resuls of each tree, we can get a majority vote. The most important parameter in the model is the number of trees, which can have a large effect on the model's performance, and a larger nubmer of trees leads to a better performance. According to the textbook, where random forest models excel are the scaliblity and ease of use. They also do not need to worry about hyper parameter values, unlike perceptrons and SVMs. Since the individual trees introduce natural randomness, a random forest model doesn't need to worry about pruning.

The model choice I would use on regular numerical data would be an SVM due to its capability of handling nonlinearly seperable data.

Question 3: Using online resources, research and find other classifier performance metrics which are also as common as the accuracy metric. Provide the MATHEMATICAL EQUATIONS for them and explain in YOUR OWN WORDS the meaning of the different metrics you found.
Note that providing math equations might invovle defining some MORE fundamental terms (define False Positive if you answer a metric that builds on that)

Other classifier Metrics: Precision,Recall [1].

True positives (TP) are values that were predicted as having a characteristic that truly have the characteristic

True negatives (TN) are values that were predicted as not having a characteristic that truly don't have that characterisitc

False Positives (FP) are values that were incorrectly predicted as having a characteristic, but don't according to the true label

False Negatives (FN) are values that were incorrectly predicted as not having a characteristic, but do indeed have it according to the true label


Precision: A measure of the true a positive prediction is. For example, it can show how many people who have a disease actually have the disase. Since it takes the ratio of true positives over the amount of calculated positives, values closer to 1 mean there are more true positives than false positives.

Equation for Precision: TP/ (TP+FP)

Recall: Measures how good your model is at predicted a class. Since precision takes the ratio of true positives over the total number of truly positive outcomes, it measures how true a prediction is. Values close to 1 mean that there are more correctly predicted outcomes than incorrectly predicted outcomes.

Equation for Recall: TP/(TP+FN)

Other classifier metrics:  NPV [2].

NPV: Measures how true a negative value is, similar to precision or PPV. Since it measures the ratio of TN over the negatively predicted values, it shows how confident we are in a value truly being negative. Values close to 1 signal that they are more true negatives than false negatives.

Equation for NPV: TN/ (TN+FN).

An ROC curve plots the true positive rate (TP/TP+FP) y across the false positive rate (FP/FP+TN) x, and plotting a model's metrics can graphically reveal the performance of the model. A perfect model would be on the upper left, since the true positive rate would be 1, while the false positive rate would be 0. By shifting different thresholds, a curve can be produced, and the closer a curve is to being diagonal, the more random the performance is.

```python
# Functions to calculate Pearson Correlation Coefficient

def mean(features):
  sum=0
  for i in range(len(features)):
    sum+=features[i]
  mean_val=sum/len(features)
  return mean_val

def std(features,mean_val):
  sumofdiff=0
  for i in range(len(features)):
    diff=features[i]-mean_val
    diffsq=(diff)**2
    sumofdiff=sumofdiff+diffsq

  std_val=(sumofdiff)**0.5
  return std_val

def numerator(mean_val1,mean_val2,features1,features2):
  sum_num=0
  for i in range(len(features1)):
    diff1=features1[i]-mean_val1
    diff2=features2[i]-mean_val2

    mult=diff1*diff2
    sum_num+=mult
  return sum_num

def PCC(numerator, std1, std2):
  pcc_val=numerator/(std1*std2)
  return pcc_val




```

```python
# Calculate Sample PCC

var1=df['GRE Score'].tolist()
var2=df['TOEFL Score'].tolist()

mean1=mean(var1)
mean2=mean(var2)

std1=std(var1,mean1)
std2=std(var2,mean2)

num12=numerator(mean1,mean2,var1,var2)
pcc1_2=PCC(num12,std1,std2)
print(pcc1_2)
```
